```python
import os
import quicksort
import numpy as np
```

```python
PATH = '../phone_dataset'
```

```python
directory_list = os.listdir(PATH)
quicksort.quickSort(directory_list)
print(directory_list)
```

```python
folders_prob = []

for example in directory_list:
        PATH_EXAMPLE = PATH + '/' + example
        NUMBER_OF_PHOTOS = len(os.listdir(PATH_EXAMPLE))
        frames_per_folder = 30
        if NUMBER_OF_PHOTOS != frames_per_folder:
            print('----------ALERT-----------')
            folders_prob.append(example)
print('Carpetas con problemas : {}'.format(folders_prob))
```
